```python
te_df = pd.read_csv("test_titanic.csv")
```

```python
te_df['Survived'] = np.nan
te_df.head(5)
```

```python
df = pd.concat([tr_df, te_df], axis=0, ignore_index=True)
df.describe()
```

```python
# it is important to know what kind of data we are working with

print(df.shape)
print(df.dtypes)
```

```python
# it is also important to know the no. of NaN values and where they lie

df.isnull().sum()
```

```python
# let us check the survival rate of the passengers

tr_df['Survived'].mean()
```

```python
# we can group data by class and view the averages for each column

tr_df.groupby('Pclass').mean()
```

```python
# changing Value for "1, 2, 3"  i.e. the different passenger classes to their respective probablities of survival ;

def getNumber(int):
    if int==1:
        return 0.652174
    elif int==2:
        return 0.486486
    elif int==3:
        return 0.256267
df["PClass"] = df["Pclass"].apply(getNumber)

#We have created a new column called 'PClass' and 
#filling it with values of eah passenger class' respective value of survival based on the values of 'Pclass' column
```

```python
# we drop the 'Name' column

df.drop('Name', axis = 1, inplace = True)
df.tail(10)
```

```python
# now we group the data by sex

sex_grouping = tr_df.groupby('Sex').mean()
sex_grouping
```

```python
# changing Value for "male, female" to their probablities of survival ; male=0.208431 and female=0.746888

def getNumber(str):
    if str=="male":
        return 0.208431
    else:
        return 0.746888
df["Gender"] = df["Sex"].apply(getNumber)

#We have created a new column called 'Sex' and 
#filling it with values of survival probablities based on the values of 'Gender' column

# we drop the 'Gender' column, since we have no use of it now

df.drop('Sex', axis = 1, inplace = True)
df.head(10)
```

```python
tr_df.groupby(['Pclass', 'Survived']).mean()
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    Survived = cols[2]
    if pd.isnull(Age):
        if Pclass == 1 and Survived == 0:
            return 42.5
        elif Pclass == 1 and Survived == 1:
            return 34.5
        elif Pclass == 2 and Survived == 0:
            return 33.5
        elif Pclass == 2 and Survived == 1:
            return 25.5
        elif Pclass == 3 and Survived == 0:
            return 27.2
        else: return 21.8
    else: return Age
```

```python
df['Age']=df[['Age','Pclass','Survived']].apply(impute_age, axis=1)
df.head(10)
```

```python
# we drop the 'Pclass' column, since we have no use of it now

df.drop('Pclass', axis = 1, inplace = True)
df.head(10)
```

```python
# checking if 'Age' column is devoid of NaN values

df.isnull().sum()
```

```python
# we drop the some more columns
df.drop(['Ticket','Cabin'], axis = 1, inplace = True)
df.head(10)
```

```python
# checking if 'Embarked' column is devoid of NaN values

df.isnull().sum()
```

```python
tr_df.groupby('Embarked').mean()
```

```python
def getNumber(str):
    if str=="C":
        return 0.563910
    elif str=="Q":
        return 0.400000
    elif str=="S":
        return 0.357438
df["Embark"] = df["Embarked"].apply(getNumber)
df.head(10)
```

```python
df['Embark'] = np.where(pd.isnull(df['Embarked']) & df['Survived']==1, 0.563910, df['Embark'])
```

```python
df['Embark'].fillna(0.357438, inplace=True)
df.isnull().sum()
```

```python
df.head(20)
```

```python
# we drop the 'Embarked' column, since we have no use of it now

df.drop('Embarked', axis = 1, inplace = True)
df.head(10)
```

```python
sibsp = df['SibSp'].values[:]
parch = df['Parch'].values[:]
fam = sibsp + parch
fam
```

```python
df['Family'] = pd.DataFrame(fam)
df.head(10)
```
